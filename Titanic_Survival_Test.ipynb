# Combining the **test and train** passengers

```python
train_index= titanic_data['PassengerId']
test_index = titanic_test_data['PassengerId']
```

```python
combined_pd = pd.concat([titanic_data, titanic_test_data]).reset_index(drop=True)
```

# Printing the Null Value of the combined data

```python
combined_pd.isnull().sum()
```

```python
y = combined_pd["Age"].value_counts()
X = y.index.values

plt.figure(figsize=(10,9))
plt.bar(X,y)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


class_column = combined_pd[["Pclass", "Age"]]
for i in range(1,4):
  plt.title("Class" > str(1))
  y = class_column[class_column["Pclass"] == i]["Age"].value_counts()
  X = y.index.values

  plt.bar(X,y)

  plt.xlabel("Age")
  plt.ylabel("Count")
  plt.show()

  midClass = class_column.groupby("Pclass").median()
  print(midClass)
```

```python
for i in range(3):
  idx = combined_pd[(combined_pd["Pclass"] == i+1) & (combined_pd["Age"].isnull())].index
  if len(idx) > 0:
    combined_pd.loc[idx, "Age"] = midClass.values[i][0]

  print(combined_pd)
```

```python
display(combined_pd[combined_pd['Fare'].isnull()])

sim_fares = combined_pd[(combined_pd['Pclass'] == 3) & (combined_pd['Embarked'] == 'S')]['Fare']
print("MidClass", sim_fares.median())

plt.hist(sim_fares)
plt.show()

combined_pd['Fare'].fillna(sim_fares.median(), inplace=True)
```

### Replaced **plt.hist** with **sns.countplot**: sns.countplot is the appropriate function from the seaborn library for visualizing the distribution of categorical data. It counts the occurrences of each category and displays them as bars. Using sns.countplot will fix the TypeError because it expects categorical data and will handle the 'Embarked' column correctly.

```python
display(combined_pd[combined_pd['Embarked'].isnull()])
sim_embark = combined_pd[(combined_pd['Pclass'] == 1) & (combined_pd['Fare'] >= 70) & (combined_pd['Fare'] <= 90)]['Embarked']

sim_embark.value_counts()

sns.countplot(x=sim_embark)
plt.show()

combined_pd['Embarked'].fillna('S', inplace=True)
```

```python
combined_pd['Cabin'].fillna('U', inplace=True)
combined_pd['Cabin'] = combined_pd['Cabin'].str[0]

print(combined_pd['Cabin'].value_counts())

idx = np.where(combined_pd['Cabin'] == 'T')[0]
combined_pd.loc[idx, 'Cabin'] = 'U'

```
